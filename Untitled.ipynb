```python
import mysql.connector
import pyodbc
import pandas


cnxn = pyodbc.connect('DRIVER={SQL Server};SERVER=localhost;PORT=1433;DATABASE=Sales;UID=sa;PWD=yourStrong(!)Password;')
cursor = cnxn.cursor()
sql = "SELECT * FROM SalesRecord"
cursor.execute(sql)
data = cursor.fetchall()
df = pandas.DataFrame.from_records(data, columns=[col[0] for col in cursor.description])
print(df)
cursor.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# if using a Jupyter notebook, includue:
%matplotlib inline


df['status'].value_counts().plot(kind='pie')
```

```python
dfg = df.groupby(['category_name'])['grant_total'].mean()

dfg.plot(kind='bar', title='Sales Revenue', ylabel='Sales',
         xlabel='Year', figsize=(6, 5))
```

```python
dfg = df.groupby(['payment_method'])['grant_total'].mean()

dfg.plot(kind='bar', title='Payment Methods', ylabel='Sales',
         xlabel='Year', figsize=(6, 5))
```
